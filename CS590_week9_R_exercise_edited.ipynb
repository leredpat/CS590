Create a new column called 'Bailey_Parent' where the value is 'Yes' if 'Bailey' was used in the development of that line, otherwise use 'No'. 
Show the first 10 rows of lines.

```r
map_condition<-function(x){
    if (grepl('Bailey', x)){
        return('YES')
    }else{
        return('NO')
    }
}

p_lines$Bailey_Parent<-lapply(p_lines$Identity.or.Parentage, map_condition)
head(p_lines, 20)
```

Create a dataframe that contains the average yield of Bailey and the average of all other lines with Bailey as the parent.

```r
map_condition2<-function(x){
    if (grepl('Bailey', x)){
        return('YES')
        
    }else if(grepl('N03081T', x)){
        return('YES')
        
    }else {return('NO')
    }
}
```

```r
peanut_db$Accession_Parent <-paste(peanut_db$NC_Accession, peanut_db$Identity.or.Parentage)
head(peanut_db, 20)
```

```r
peanut_db$Bailey<-sapply(peanut_db$Accession_Parent, map_condition2)
head(peanut_db, 50)
```

```r
peanut_Bailey<-subset(peanut_db, Bailey=='YES')
```
