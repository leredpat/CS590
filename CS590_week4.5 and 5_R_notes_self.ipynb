```r
peanut.weight <- data.frame(
    name = I(c('Bailey','Bailey II','Wynne','Emery','Sullivan', 'Gregory')),
    weight = c(89.4, 92.3, 105.2, 103.1, 99.6, 89.1))
peanut.skins <- data.frame(
    name = I(c('Bailey','Bailey II','Wynne','Emery','Sullivan','Brantley')),
    skin = c('Tan', 'Tan', 'Pink', 'Pink', 'Tan', 'Red'))
```

```r
peanut.weight
```

```r
peanut.skins
```

```r
#merge ()
#outerjoin
#by.x tells you to which you are joining/merging the datasets by
peanut<-merge(peanut.weight,peanut.skins, by.x='name',by.y='name',all=TRUE)

#leftjoin on arrangement that we have = use all.x
#rightjoin = all.y
#interjoin/outer is the default ('all' is missing from statement)
```

```r
peanut
```

```r
#working with missing data
#any()
#is.na()

#if there is any missing data (data with values na) within 'peanut'
any(is.na(peanut))
```

```r
any(is.na(peanut$skins))
```

```r
any(is.na(peanut$weight))
```

```r
peanut[is.na(peanut$weight),]
```

```r
#conditional statement for not equal to, 
#this statement pulls out everything that does not have an 'NA' in it (essentially will drop 'Brantley')

peanut[!is.na(peanut$weight),]
```

```r
#will redefine missing Na values as the mean value for that variable
#na.rm = finds and replaces the NA values without including them in the mean calculation
peanut$weight[is.na(peanut$weight)]<-mean(peanut$weight,na.rm=TRUE)
```

```r
peanut
```

```r
peanut$skin[is.na(peanut$skin)] <- 'Tan'
```

```r
peanut
```
