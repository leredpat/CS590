```r
# Install ggplot2
install.packages("ggplot2")
```

```r
# Import ggplot
library(ggplot2)
```

```r
# Building Grammar Layers for ggplot2 Figures

# Data + Aesthetics
ggplot(data=mtcars, aes(x=mpg, y=hp))
#not plotting bc we havent defined geometry!!
```

```r
# Data + Aesthetics + Geometries
ggplot(data=mtcars, aes(x=mpg, y=hp)) + geom_point()
#geom_point=basic scatter
```

```r
# Data + Aesthetics + Geometries + Facets
ggplot(data=mtcars, aes(x=mpg, y=hp)) + geom_point() + facet_grid(mtcars$cyl)
#use facet to look at different reps - in this case looks at it based off of cylinder
```

```r
# Data + Aesthetics + Geometries + Facets + Statistics
ggplot(data=mtcars, aes(x=mpg, y=hp)) + geom_point() + facet_grid(mtcars$cyl) + stat_smooth(method = loess, formula = y ~ x)
```
